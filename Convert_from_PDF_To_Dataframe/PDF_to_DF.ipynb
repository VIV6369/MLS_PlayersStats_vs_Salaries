```python
import pdfplumber
import pandas as pd

with pdfplumber.open("2021-MLSPA-Fall-Salary-release.pdf") as pdf:
    # First we need to extract columns so later on they dont provide IndexError while doing pd.concat
    page1 = pdf.pages[0]
    # Extract the table from the first page
    table1 = page1.extract_table()
    # Defining columns
    columns=table1[0]
    all_tables = []
    for i in range(0,len(pdf.pages)):
        # Select pages of the PDF
        page = pdf.pages[i]
        # Extract the table from pages
        table = page.extract_table()
        # Convert the extracted table data from pages into a Pandas dataframe
        df = pd.DataFrame(table, columns=columns)
        all_tables.append(df)
    
    # Concatenate the two dataframes into a single Pandas dataframe
    df = pd.concat(all_tables).reset_index(drop=True)

# Remove duplicate header
df.drop(0, axis=0, inplace=True)
# Reset index
df = df.reset_index(drop=True)
df   
```
