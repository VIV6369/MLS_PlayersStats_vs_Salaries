#### Dataframe split

For more precise results we have to split dataframe into 4 categories:

    * Forwards
    Players who play the furthest up the pitch therefore most responsible for scoring goals as well as assisting them.
    
    * Midfielders
    Midfielders often go across boundaries, with mobility and passing ability, they are often referred to as deep-lying midfielders, play-makers, box-to-box midfielders, or holding midfielders. Midfielders are assigned to disrupt the opposing team's attacks, while others may be tasked with creating goals, or have equal responsibilities between attack and defence. They are crucial and core of any team.
    
    * Defenders
    Players whose primary role is to stop attacks during the game and prevent the opposition from scoring.
    
    * Goalkeepers
    The goalkeeper's main role is to stop the opposing team from scoring (putting the ball over the goal-line of the goal). This is accomplished by having the goalkeeper move into the trajectory of the ball to either catch it or direct it further from the vicinity of the goal line. 
    
Since players have different roles on the field, we should compare their stats accordingly. Only within same category.


```python
# Break down by position
forwards = stats[stats.position == 'Forward']
midfielders = stats[stats.position == 'Midfielder']
defenders = stats[stats.position == 'Defender']
goalkeepers = stats[stats.position == 'Goalkeeper']

print(len(forwards), len(midfielders), len(defenders), len(goalkeepers))
```

#### Sample size

Any sort of statistical analysis requires a minimum sample size. For example, a player who played only one game per season could have one shot on target and score one goal, resulting in a goal_conversion_per of 100%. On paper, this player would be the most efficient forward in the league, but in reality, the sample size is simply not big enough. To avoid such scenarios, we need to increase our sample size. To do so, we will filter out players who have played fewer than 765 minutes per season. 765 minutes is a 25% minimum threshold since the regular season has a total of 3060 regulation minutes (34 x 90).

```python
# Filter out players with small sample size
forwards = forwards[forwards.mins >= 765].reset_index(drop=True)
midfielders = midfielders[midfielders.mins >= 765].reset_index(drop=True)
defenders = defenders[defenders.mins >= 765].reset_index(drop=True)
goalkeepers = goalkeepers[goalkeepers.mins >= 765].reset_index(drop=True)
```

#### Forwards

To compare statistics for forwards, we will need to take a closer look at mainly attacking numbers, since it's their main role, along with some ball possession stats as well. Here they are:

    - goal_conversion_per
    - accurate_pass_per
    - accurate_cross_per
    - accurate_fwd_zone_pass_per
    - tackle_success_per
    - accurate_shooting_per

The following stats are whole numbers for the season, but we need to adjust them based on the minutes spent on the field for each player or as a success percent:

    - assist
    - successful_dribble
    - ball_recovery
    - duel_won
    - interception
    - accurate_layoffs  
    
Most of the time penalty kikcs are done by designated shooters. Which means for majority of players this feature is not available, therefore we should adjust it relatevly to penalty kicks taken not minutes on the field. 

```python
# ADJUST NEEDED TO FORWARS STATS + LAYOFF
# Create minutes adjusted stats
midfielders['goal_conversion_per_min'] = midfielders['goal_conversion_per'] / midfielders['mins']
midfielders['accurate_pass_per_min'] = midfielders['accurate_pass_per'] / midfielders['mins']
midfielders['accurate_cross_per_min'] = midfielders['accurate_cross_per'] / midfielders['mins']
midfielders['accurate_fwd_zone_pass_per_min'] = midfielders['accurate_fwd_zone_pass_per'] / midfielders['mins']
midfielders['tackle_success_per_min'] = midfielders['tackle_success_per'] / midfielders['mins']
midfielders['accurate_shooting_per_min'] = midfielders['accurate_shooting_per'] / midfielders['mins']
midfielders['goals_min'] = midfielders['goals'] / midfielders['mins']
midfielders['assists_min'] = midfielders['assist'] / midfielders['mins']
midfielders['ball_recovery_min'] = midfielders['ball_recovery'] / midfielders['mins']
midfielders['interception_min'] = midfielders['interception'] / midfielders['mins']
midfielders['total_through_ball_min'] = midfielders['total_through_ball'] / midfielders['mins']
midfielders['won_corners_min'] = midfielders['won_corners'] / midfielders['mins']
# Calculate percetanges and minutes adjusted for those we don't have yet
# Dribbles
midfielders['total_dribbles'] = midfielders['successful_dribble'] + midfielders['failed_dribble']
midfielders['successful_dribble_per'] = midfielders['successful_dribble'] / midfielders['total_dribbles'] 
midfielders['successful_dribble_per_min'] = midfielders['successful_dribble_per'] / midfielders['mins']
# Back zone pass
midfielders['total_back_zone_pass'] = midfielders['accurate_back_zone_pass'] + midfielders['failed_back_zone_pass']
midfielders['accurate_back_zone_pass_per'] = midfielders['accurate_back_zone_pass'] / midfielders['total_back_zone_pass'] 
midfielders['accurate_back_zone_pass_per_min'] = midfielders['accurate_back_zone_pass_per'] / midfielders['mins']
# Duels
midfielders['duel_won_per'] = midfielders['duel_won'] / midfielders['duel']
midfielders['duel_won_per_min'] = midfielders['duel_won_per'] / midfielders['mins']
# Short pass
midfielders['total_short_pass'] = midfielders['successful_short_pass'] + midfielders['failed_short_pass']
midfielders['successful_short_pass_per'] = midfielders['successful_short_pass'] / midfielders['total_short_pass']
midfielders['successful_short_pass_per_min'] = midfielders['successful_short_pass_per'] / midfielders['mins']
# Long Balls
midfielders['total_long_balls'] = midfielders['accurate_long_balls'] + midfielders['failed_long_balls']
midfielders['accurate_long_balls_per'] = midfielders['accurate_long_balls'] / midfielders['total_long_balls']
midfielders['accurate_long_balls_per_min'] = midfielders['accurate_long_balls_per'] / midfielders['mins']
# Fifty-fifty
midfielders['successful_fifty_fifty_per'] = midfielders['successful_fifty_fifty'] / midfielders['fifty_fifty']
midfielders['successful_fifty_fifty_per_min'] = midfielders['successful_fifty_fifty_per'] / midfielders['mins']
# Penalty goal percent
midfielders['penalty_goal_per'] = midfielders['penalty_goal'] / midfielders['pen_taken']
midfielders['penalty_goal_per_min'] = midfielders['penalty_goal_per'] / midfielders['mins']
# Fill NaN with 0 for players who doesn't shoot penalties
midfielders['penalty_goal_per'] = midfielders['penalty_goal_per'].fillna(0)
midfielders['penalty_goal_per_min'] = midfielders['penalty_goal_per_min'].fillna(0)

forwards.head()
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

# Instantiate standard scaler
scaler = StandardScaler()
# Fit transform data
array_nums_scaled = scaler.fit_transform(fw_nums)
# Creating scaled df
fw_nums_scaled = pd.DataFrame(array_nums_scaled, columns=columns)
print(fw_nums_scaled.head())
```

```python
# Split data into test and train dataset
from sklearn.model_selection import train_test_split

# Select predictors
X = fw_nums_scaled[['height, cm', 'weight, kg',
       'goal_conversion_per', 'accurate_pass_per',
       'accurate_cross_per', 'accurate_fwd_zone_pass_per', 'tackle_success_per',
       'accurate_shooting_per', 'assists_per_min', 'successful_dribble_per_min',
       'ball_recovery_per_min', 'duel_won_per_min', 'interception_per_min',
       'accurate_layoffs_per_min', 'penalty_goal_per']]
# Select target variable
y = fw_nums_scaled[['2022 guar. comp.']]

# Split data
x_train, x_test, y_train, y_test = train_test_split(X,y, train_size=0.8, test_size=0.2)
```

```python
# Fit model
from sklearn.linear_model import LinearRegression

# Ordinary Least Squares method
ols_all = LinearRegression()
# fit model
ols_all.fit(x_train,y_train)
# predict compensation on test data
comp_predict_all = ols_all.predict(x_test)
# R2 score (Residual Analysis)
print('All features Train score:')
print(ols_all.score(x_train,y_train))
print('All features Test score:')
print(ols_all.score(x_test,y_test))
# All features coef
print("All features coefficients:")
print(ols_all.coef_)
```

#### Distribution

In order to be able to implement prediction model we need to make sure that data is normaly distributed. Let's check that.

```python
# Distribution of compensation
# Set our variable
salary = forwards['2022 guar. comp.']
# Graph our compensation readings
plt.hist(salary, bins=70)
plt.xticks(rotation=45)
plt.title('Number of players by Compensation', size=16)
plt.ylabel('Number of players',size=12)
plt.xlabel('Guaranteed compensation',size=12)
plt.show()
plt.clf()
```

As we can see from the graph above compesation is not "normaly" distributed it is right-skewed histogram. With some outliers which are so called designated players. Since we want to predict compensation we have to use this feature, but model will struggle to pick up patterns for the players with high compensation (above 2 mil.) since there are only few of them and majority of players are under 2 mil. 

In order to address this issue we will try to scale this feature by using log transformation.

```python
# Set our variable
log_salary = np.log(forwards['2022 guar. comp.'])
# Graph our compensation readings
plt.hist(log_salary, bins=70)
plt.xticks(rotation=45)
plt.title('Logarithm of Compensation', size=16)
plt.show()
plt.clf()
```

Now it looks more like "normal" distribution. Now we can try log-transformed compensation as target variable and see if model will perform better.

#### Elastic Net


```python
from sklearn.linear_model import ElasticNet

# Scale log_salary 
log_salary_reshaped = np.array(log_salary).reshape(-1,1)
log_salary_scaled = scaler.fit_transform(log_salary_reshaped)

# Select predictors
X = fw_nums_scaled[['height, cm', 'weight, kg',
       'goal_conversion_per', 'accurate_pass_per',
       'accurate_cross_per', 'accurate_fwd_zone_pass_per', 'tackle_success_per',
       'accurate_shooting_per', 'assists_per_min', 'successful_dribble_per_min',
       'ball_recovery_per_min', 'duel_won_per_min', 'interception_per_min',
       'accurate_layoffs_per_min', 'penalty_goal_per']]

# Select target variable
y = log_salary_scaled

# Split data
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)

# Elastic Net method
enet = ElasticNet(alpha=0.1, l1_ratio=0.5)
# Fit model
enet.fit(x_train, y_train)
# Predict compensation on test data
comp_predict_log = enet.predict(x_test)

# R2 score (Residual Analysis)
print('Y-log features Train score:')
print(enet.score(x_train,y_train))
print('Y-log features Test score:')
print(enet.score(x_test,y_test))

# Elastic Net coefficients
print("Y-log features coefficients:")
print(enet.coef_)
```
