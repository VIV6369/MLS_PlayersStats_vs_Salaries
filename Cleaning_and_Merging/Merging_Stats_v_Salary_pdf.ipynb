# Salary report

Salary report will be gathered from MLS Players Association (http://mlsplayers.org/resources/salary-guide). 

MLSPA publishes it yearly starting from 2007. It can be downloaded as pdf file. 

The data from pdf file can be loaded into DataFrame and later on added to the Players Statisctics DataFrame.

```python
# Load pdf reader library
import pdfplumber
import pandas as pd

with pdfplumber.open("2021-MLSPA-Fall-Salary-release.pdf") as pdf:
    # First we need to extract columns so later on they dont provide IndexError while doing pd.concat
    page1 = pdf.pages[0]
    # Extract the table from the first page
    table1 = page1.extract_table()
    # Defining columns
    columns=table1[0]
    all_tables = []
    for i in range(0,len(pdf.pages)):
        # Select pages of the PDF
        page = pdf.pages[i]
        # Extract the table from pages
        table = page.extract_table()
        # Convert the extracted table data from pages into a Pandas dataframe
        data = pd.DataFrame(table, columns=columns)
        all_tables.append(data)
    
    # Concatenate the two dataframes into a single Pandas dataframe
    salary = pd.concat(all_tables).reset_index(drop=True)
    
salary.drop(0, axis=0, inplace=True)
salary = salary.reset_index(drop=True)
salary
```

#### Clean data

Salary dataframe needs to be cleaned and transformed before any analysis.

```python
# Clean salary amounts and convert them to numerical value
# Clean Base Salay column
salary[salary.columns[4]] = salary[salary.columns[4]].replace('[\$,]', '', regex=True)
salary[salary.columns[4]] = pd.to_numeric(salary[salary.columns[4]])
# Clean Guaranteeed Comp. column
salary[salary.columns[5]] = salary[salary.columns[5]].replace('[\$,]', '', regex=True)
salary[salary.columns[5]] = pd.to_numeric(salary[salary.columns[5]])
# Check outcome
print(salary[salary.columns[4]].head())
print(salary[salary.columns[5]].head())

# Some last names have "n" in front of them let's correct it.
salary['Last Name'] = salary['Last Name'].str.replace('^n', '', regex=True)

# Creating full name column
salary['name'] = salary['First Name'] + ' ' + salary['Last Name'] 
print(salary.name.head())
```

```python
# Remove diacritical marks from the 'name' column
from unidecode import unidecode

salary['name'] = salary['name'].apply(unidecode)
salary['name'] = salary['name'].str.title()
```

Find names with different spelling for the same players and replace them with the ones in salary df

```python
# This code would compare the last names of the players in salary[not_in_df]['name'] and df['name'], 
# and replace the df['name'] value with the value from salary[not_in_df]['name'] if last name and club abbreviation matches.
for i in salary[not_in_df]['name']:
    last_name_salary = i.split()[-1]
    for j in df['name']:
        last_name_df = j.split()[-1]
        if last_name_salary == last_name_df and salary[salary['name'] == i]['club_abb'].values == df[df['name'] == j]['club_abb'].values:
            df['name'] = df['name'].replace(j, i)
```
